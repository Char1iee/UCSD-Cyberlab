```python
import pandas as pd

import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.utils import shuffle
```

```python
df = pd.read_csv('/home/jovyan/WUSTL-IIoT/dataset/original/wustl_iiot_2021.csv', low_memory=False) 
```

```python
df.head(5)
```

```python
df.dtypes
```

```python
drop_columns = ['StartTime', 'LastTime', 'SrcAddr', 'DstAddr', 'sIpId', 'dIpId', 'Target']

df.drop(drop_columns, axis=1, inplace=True)

df.dropna(axis=0, how='any', inplace=True)

df.drop_duplicates(subset=None, keep="first", inplace=True)

df = shuffle(df)

df.isna().sum()

print(df['Traffic'].value_counts())
```

```python
df.dtypes
```

```python
np.unique(df.Traffic)
```

```python
target = df['Traffic']
df = df.drop(columns = ['Traffic'])
```

```python
encoder = LabelEncoder()
target = encoder.fit_transform(target.values.reshape(-1, 1))
target
```

```python
scaler = MinMaxScaler()
df = scaler.fit_transform(df)
```

```python
pd.DataFrame(df).describe()
```

```python
x_train, x_test, y_train, y_test = train_test_split(df, target, test_size=0.2, random_state=21)
```

```python
np.save("/home/jovyan/WUSTL-IIoT/dataset/preprocessed/x_train.npy", x_train)
np.save("/home/jovyan/WUSTL-IIoT/dataset/preprocessed/x_test.npy", x_test)
np.save("/home/jovyan/WUSTL-IIoT/dataset/preprocessed/y_train.npy", y_train)
np.save("/home/jovyan/WUSTL-IIoT/dataset/preprocessed/y_test.npy", y_test)
```
